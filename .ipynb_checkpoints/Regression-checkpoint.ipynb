## Lineare Regression

Lineare Regression für count
Key steps:
1. exploratory data analysis
2. train/test split
3. feature engineering
4. train a model
5. cross-validation
6. optimize the model iteratively, select features, 
    try different regressors 
    (e.g. Linear Regression, Random Forest Regressor, SVR)
7. calculate a test score when you are done

```python
#module
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
#train/test split
X_df_sams = df_sams[['temp','humidity','season', 'hours']]
y_df_sams = df_sams['count']
X_train, X_test, y_train, y_test = train_test_split(X_df_sams,y_df_sams,test_size=0.2, random_state=42)
m=LinearRegression()
```

### train a model

```python
from sklearn.preprocessing import PolynomialFeatures
# interaktionsterm mit season
poly = PolynomialFeatures(interaction_only=True)
poly_features = poly.fit_transform(X_train[['temp','humidity','hours', 'season']])

#linear regression
m.fit(poly_features, y_train)
ypred_exp = m.predict(poly_features)
```

```python
poly.get_feature_names()
```

```python
m.score(poly_features, y_train)
```

#### Logistic Regression casual vs. registered

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
X_log = df[['season', 'hours','weekday']]
y_log = df['registered']

X_train_log, X_test_log, y_train_log, y_test_log = train_test_split(X_log, y_log, test_size=.25, random_state=0)

clf = LogisticRegression(random_state=0)

# training
clf.fit(X_train_log, y_train_log)
# predicting with train
y_pred_train_log = clf.predict(X_train_log)

# now testing
y_pred_test_log = clf.predict(X_test_log)

X_train.shape, y_train.shape,X_test.shape, y_test.shape
clf.score(X_train_log,y_train_log),clf.score(X_test_log,y_test_log)

accuracy_score(y_test_log, y_pred_test_log),accuracy_score(y_train_log, y_pred_train_log)
```
