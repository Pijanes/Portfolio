#### Create an animated scatterplot

```python
df6
```

```python
for i in range(1950,2016):
    df6=df6s[i]
    df6=df6.unstack(1)

    cmap = plt.get_cmap('tab20', lut=len(df6)). colors
    df6.plot.scatter('fertility', 'lifeexp', s=3, c=cmap)
    plt.legend(loc=0)
    plt.axis((0,10,0,90))
    filename=str("lifeexp_"+str(i)+ ".png")
    plt.savefig(filename)
    
    #save as lifeexp_[i].png 
```

```python
import imageio
images = []

for i in range(1950, 2016):
    filename = 'lifeexp_{}.png'.format(i)
    images.append(imageio.imread(filename))

imageio.mimsave('output.gif', images, fps=20)
#for continent, continent_df in g1:
```
